IMPORT THE REQUIRED LIBRARIES FOR THIS PROJECT

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas DATAFRAME
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# first 10 rows of the data set
credit_card_data.head(10)
```

```python
#checking the last five rows of the dataset
credit_card_data.tail(5)
```

```python
# information about the particular data set
credit_card_data.info()
```

```python
#checking for missing values from the data set from each column
credit_card_data.isnull().sum()
```

```python
#checking for the distribution of the legit and fraud transaction
credit_card_data['Class'].value_counts()
```

This data set is highly unbalanced

0 -> NORMAL TRANSACTION

1-> FRAUDALENT TRANSACTION

```python
#separating the data for analysis
legit_1 = credit_card_data[credit_card_data.Class == 0]
fraud_1 = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit_1.shape)
print(fraud_1.shape)
```

```python
#statistical measures of the data
legit_1.Amount.describe()
```

```python
#statistical measures of the data for fraudalent rows
fraud_1.Amount.describe()
```

```python
# comparing the values of both the transactions
credit_card_data.groupby('Class').mean()
```

UNDER-SAMPLING

Build a Sample DataSet containing similar Distribution of Normal Transactions and Fraudalent Transactions

Number of Fraudalent Transactions -> 492

```python
legit_sample = legit_1.sample(n=492)
```

Concatenating the two Datasets

```python
new_data_set = pd.concat([legit_sample ,fraud_1],axis=0)
```

```python
new_data_set.head()
```

```python
new_data_set.tail()
```

```python
#counting the data in concatenated file for analysis
new_data_set['Class'].value_counts()
```

```python
new_data_set.groupby('Class').mean()
```

SPLITTING THE DATA INTO FEATURES AND TARGETS

```python
X = new_data_set.drop(columns='Class',axis=0)
Y = new_data_set['Class']
print(X)
```

```python
print(Y)
```

SPLITTING THE DATA INTO TRAINING AND TESTING DATA


```python
X_Train , X_Test,Y_Train , Y_Test = train_test_split(X,Y,test_size = 0.2 , stratify = Y,random_state = 2)
```

```python
print(X.shape,X_Train.shape,X_Test.shape)
```

MODEL TRAINING

LOGISTIC REGRESSION(for binary classification)

```python
model_1 = LogisticRegression()
```

```python
#training the logistic regrssion model with training data
model_1.fit(X_Train,Y_Train)
```

EVALUATION OF THE DEVELOPED MODEL BASED ON ACCURACY SCORE

```python
#accuracy score based on training data
X_Train_prediction = model_1.predict(X_Train)
training_data_accuracy  = accuracy_score(X_Train_prediction,Y_Train)
print('ACCURACY ON TRAINING DATA : ',training_data_accuracy)
```

```python
#ACCURACY ON TEST DATA
X_Test_prediction = model_1.predict(X_Test)
test_data_accuracy = accuracy_score(X_Test_prediction,Y_Test)
print("ACCURACY SCORE ON TEST DATA:",test_data_accuracy)
```
